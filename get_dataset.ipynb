```python
from datasets import load_dataset

# Charger le dataset
dataset = load_dataset("dbarbedillo/SMS_Spam_Multilingual_Collection_Dataset")

# Sélectionner uniquement les colonnes des textes en français et label (ham/spam)
fr_data = dataset["train"].map(lambda x: {"text": x["text_fr"], "label": x["labels"]})

# Retirer les entrées vides
fr_data = fr_data.filter(lambda x: x["text"] is not None and x["text"].strip() != "")

# Vérifier
print(len(fr_data))

# Afficher les 5 premiers exemples
for i in range(5):
    print(fr_data[i])
    print("-" * 50)

```
